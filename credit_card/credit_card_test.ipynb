```python
import pandas as pd
credit = pd.read_csv('drive/MyDrive/venv/0_pkg/kaggle/credit_card_client/UCI_Credit_Card.csv')
credit.info()
```

```python
c_X = credit.drop('default.payment.next.month', axis=1)
c_y = credit['default.payment.next.month']
c_X.shape
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()

hyper_param = {'n_estimators':[100, 200, 300, 400],
               'max_depth':[5, 6, 7, 8],
               'min_samples_split':[2, 3, 4, 5, 6]}
```

```python
gscv = GridSearchCV(model, param_grid=hyper_param,
                    scoring='accuracy', refit=True,
                    cv=3, return_train_score=True)
```

```python
gscv.fit(c_X, c_y)
print(gscv.best_params_)
print(gscv.best_score_)
print(gscv.best_estimator_)
```
